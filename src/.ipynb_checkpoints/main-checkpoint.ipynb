### Modelo ARIMA

### Comprobación de la estacionariedad

```python
# función que incluye gráfica y test de Dickey-Fuller

def test_stationarity(timeseries):
    
    # gráfica                                       
    rolmean = timeseries.rolling(window=30).mean() 
    rolstd = timeseries.rolling(window=30).std()     
        
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)    
    
    # Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])

    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
plt.figure(figsize=(16,8)) 
test_stationarity(train['Count'])
```

```python
# En la gráfica se aprecia claramente que la media crece; y la varianza también lo hace, aunque más 
# suavemente
```

```python
# El test de Dickey-Fuller es más esclarecedor. El estadístico es mayor que cualquiera de los valores 
# críticos, luego no se rechaza la hipótesis nula y se considera que la serie no es estacionaria.
```

#### Transformaciones de la serie

```python
# estabilización de la varianza mediante transformación logarítmica

train_log = np.log(train.Count) 
plt.figure(figsize=(16,8)) 
test_stationarity(train_log)
```

```python
# eliminación de la tendencia mediante diferenciación 1 día 

train_log_diff = train_log.diff().dropna()
plt.figure(figsize=(16,8)) 
test_stationarity(train_log_diff)
```

```python
# eliminación de la estacionalidad mediante diferenciación 7 días 

train_log_diff_diff7 = train_log_diff.diff(7).dropna()  
plt.figure(figsize=(16,8)) 
test_stationarity(train_log_diff_diff7)
```

```python
# Tras aplicar las transformaciones, el estadístico es menor que el valor crítico (1%), por lo que se 
# rechaza la hipótesis nula y se considera que la serie es estacionaria. Tenemos un nivel de confianza del
# 99%.
```

### Determinación automatizada de los parámetros del modelo SARIMAX

```python
# fijación del valor de los parámetros entre 0 y 2

p = d = q = range(0, 2)
```

```python
# generación de todas las posibles combinaciones de p, d y q 

pdq = list(itertools.product(p, d, q))
```

```python
# generación de todas las posibles combinaciones de los parámetros P, D y Q dado m 

m = 7
seasonal_pdq = [(x[0], x[1], x[2], m) for x in list(itertools.product(p, d, q))]
```

```python
results_list = []    
for param in pdq:
    for param_seasonal in seasonal_pdq:  
        try:            
            mod = SARIMAX(train_log,
                  order=param,
                  seasonal_order=param_seasonal,
                  enforce_stationarity=False,
                  enforce_invertibility=False)

            result = mod.fit()
            results_list.append([param, param_seasonal, result.aic])            
        except:
            continue
            
result_min = results_list[0]  
for result in results_list:
    if result[2] < result_min[2]:
        result_min = result
print('SARIMAX{}{} - AIC:{}'.format(result_min[0], result_min[1], result_min[2]))       
```
